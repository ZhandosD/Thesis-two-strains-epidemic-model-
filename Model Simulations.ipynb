```python
# library implementation

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# model definition

def seir_model(t, y, params):
    S, E_d, E_o, I_d, I_o, R_d, R_o, V = y
    (beta_d, beta_o, sigma_d, sigma_o, gamma_d, gamma_o, mu_d, mu_o, 
     nu, omega, phi, epsilon_d, epsilon_o, eta_d_o, Lambda) = params
    
    u_t = 0.5  # Example intervention function (constant for now)
    
    dS_dt = - (1 - u_t) * (beta_d * I_d + beta_o * I_o) * S - (nu + phi) * S + omega * (V + R_d + R_o) + Lambda
    dE_d_dt = (1 - u_t) * beta_d * I_d * (S + epsilon_d * V) - (sigma_d + phi) * E_d
    dE_o_dt = (1 - u_t) * beta_o * I_o * (S + eta_d_o * R_d + epsilon_o * V) - (sigma_o + phi) * E_o
    dI_d_dt = sigma_d * E_d - (gamma_d + mu_d + phi) * I_d
    dI_o_dt = sigma_o * E_o - (gamma_o + mu_o + phi) * I_o
    dR_d_dt = gamma_d * I_d - (1 - u_t) * eta_d_o * I_o * R_d - (omega + phi) * R_d
    dR_o_dt = gamma_o * I_o - (omega + phi) * R_o
    dV_dt = nu * S - (omega + epsilon_d + epsilon_o + phi) * V
    
    return [dS_dt, dE_d_dt, dE_o_dt, dI_d_dt, dI_o_dt, dR_d_dt, dR_o_dt, dV_dt]
```

```python
# Initial conditions

S0, E_d0, E_o0, I_d0, I_o0, R_d0, R_o0, V0 = 0.9, 0.2, 0.2, 0.3, 0.2, 0, 0, 0  # Example values
initial_conditions = [S0, E_d0, E_o0, I_d0, I_o0, R_d0, R_o0, V0]
```

```python
# Parameters

params = (
    0.3, 0.4,    # beta_d, beta_o
    1/5, 1/4,    # sigma_d, sigma_o (inverse latency periods)
    1/10, 1/8,   # gamma_d, gamma_o (recovery rates)
    0.01, 0.005, # mu_d, mu_o (disease-induced death rates)
    0.005, 0.02,  # nu (vaccination rate), omega (waning immunity rate)
    0.0001,      # phi (natural death rate)
    0.6, 0.5,    # epsilon_d, epsilon_o (vaccine efficiency)
    0.3,         # eta_d_o (cross-immunity coefficient)
    0.0001       # Lambda (birth rate)
)
```

```python
# Time span

t_span = (0, 365)  # Simulate for one year
t_eval = np.linspace(t_span[0], t_span[1], 365)
```

```python
# Solve the system

solution = solve_ivp(seir_model, t_span, initial_conditions, args=(params,), t_eval=t_eval)
```

```python
# Plot results

plt.figure(figsize=(10, 6))
plt.plot(solution.t, solution.y[0], label='Susceptible')
plt.plot(solution.t, solution.y[1], label='Exposed (Delta)')
plt.plot(solution.t, solution.y[2], label='Exposed (Omicron)')
plt.plot(solution.t, solution.y[3], label='Infected (Delta)')
plt.plot(solution.t, solution.y[4], label='Infected (Omicron)')
plt.plot(solution.t, solution.y[5], label='Recovered (Delta)')
plt.plot(solution.t, solution.y[6], label='Recovered (Omicron)')
plt.plot(solution.t, solution.y[7], label='Vaccinated')
plt.xlabel('Days')
plt.ylabel('Population')
plt.legend()
plt.title('SEIR Model with Two Strains and Vaccination')
plt.grid()
plt.show()
```
